```python
df.dropna(how='all', subset=['END_DATE','CATEGORY','START','STOP','PURPOSE'], inplace=True)
df.isna().sum()
```

```python
(df['PURPOSE'].isna().sum() / df.shape[0])*100 # percent of null data in purpose column
```

```python
#identifying the top start-stop (locations) pairs that account for the most mileage on avg
df.groupby(['START','STOP'])['MILES'].mean().sort_values(ascending=False).head()
```

```python
# duration of drives
df['DURATION'] = df.END_DATE - df.START_DATE
df['DURATION'] = df.DURATION.dt.total_seconds()/60  # convert duration to minutes
df.sample(5)
```

```python
def rounds(x):
  if x.START == x.STOP:
    return 'yes'
  return 'no'
```

```python
df.apply(rounds,axis=1) # assign to a new col name to add to df
```

```python
lst = []
for i,j in zip(df.START,df.STOP):
  if i == j:
    lst.append('yes')
  else:
    lst.append('no')
```

```python
df['ROUND_TRIP'] = lst
df.sample(5)
```

```python
sns.countplot(df.ROUND_TRIP)
```

```python
# purpose vs miles/duration
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(data=df,x=df.PURPOSE,y=df.MILES)
plt.xticks(rotation=45)
plt.subplot(1,2,2)
sns.boxplot(data=df,x=df.PURPOSE,y=df.DURATION)
plt.xticks(rotation=45)
```

there is a relationship between miles and minutes (obvio), as shown in the graphs

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
# months with most bookings
df['MONTH'] = pd.DatetimeIndex(df.START_DATE).month
df.tail(5)
```

```python
def func(x):
  d = {
      1: 'Jan',
      2: 'Feb',
      3: 'Mar',
      4: 'Apr',
      5: 'May',
      6: 'Jun',
      7: 'Jul',
      8: 'Aug',
      9: 'Sep',
      10: 'Oct',
      11: 'Nov',
      12: 'Dec'
       }
  return d[x]
```

```python
df['MONTH'] = df['MONTH'].apply(func)
df.tail(5)
```

```python
sns.countplot(df.MONTH)
```

```python
sns.countplot(df.MONTH, order=df.MONTH.value_counts().index)
```

more booking around december - christmas

```python
sns.lineplot(x=df.DURATION,y=df.MILES)
```

```python
# purpose and category
plt.figure(figsize=(10,3))
sns.countplot(data=df,x='PURPOSE',hue='CATEGORY',dodge=False)
plt.xticks(rotation=45)
```

### Conclusions

1. **Higher number of trips observed in December.**
2. **Business trips outnumber personal trips.**
3. **Trip purposes like Commute, Moving, and Charity are exclusively personal.**
4. **Trips between offices and airports are exclusively for business.**
5. **Seasonal patterns, such as festivals and national holidays, influence booking trends.**
6. **Majority of bookings are for meetings, meals/entertainment, and temporary site visits.**
7. **More miles usually mean longer duration, but shorter trips can still take longer due to traffic or other delays.**
8. **Round trips are present in the dataset.**
